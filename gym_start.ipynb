# OpenAI Gym Starting

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
env = gym.make('LunarLander-v2')
env.seed(1)
```

```python
env.reset()
actions = []
rewards = []
running_reward = 0
for steps_episode in range(1000):
    action = env.action_space.sample()
    actions.append(action)
    env.render()
    state, reward, done, _ = env.step(action)
    rewards.append(reward)
    if done:
        break
```

### Plots

```python
plt.plot(actions)
```

```python
plt.plot(rewards)
```
